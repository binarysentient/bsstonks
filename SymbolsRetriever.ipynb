```python
import urllib.request, json, csv, os

api_token = os.environ['IEX_KEY']
data_path = os.environ['BSSTONKS_DATA_LOCATION']

#  
#   Response Format
#   {
#     "symbol": "012-LN",
#     "exchange": "LON",
#     "exchangeSuffix": "LN",
#     "exchangeName": "LONDON STOCK EXCHANGE",
#     "name": "012 Golden Lines",
#     "date": "2021-06-23",
#     "type": "cs",
#     "iexId": null,
#     "region": "GB",
#     "currency": "GBP",
#     "isEnabled": true,
#     "figi": "BBG000BGF0H6",
#     "cik": null,
#     "lei": null
#   }
    
# Region codes of important trading - source https://stockmarketmba.com/globalstockexchanges.php,
# Leaving India out of here as it is covered in another script
global_trading_regions = ['US', # United States
                          'GB', # Great Britain
                          'DE', # Germany
                          'HK', # Hongkong
                          'NL', # Netherland
                          'AE', # United Arab Emirated
                          'KS', # South Korea
                          'CN', # China
                          'JP', # Japan
                          'CA', # Canada
                          'FR', # France
                          'CH', # Switzerland
                          'AU', # Australia
                         ]

with open(data_path+'/global_symbols_iex_2.csv', 'w', newline = '') as csvfile:
    csv_writer = csv.writer(csvfile, delimiter = ',')
    csv_writer.writerow(['Symbol', 'Company Name', 'Region', 'Currency', 'Exchange' ])
    for region in global_trading_regions:
        print('Will load symbols for {}'.format(region))
        with urllib.request.urlopen("https://cloud.iexapis.com/beta/ref-data/region/{}/symbols?token={}".format(region, api_token)) as url:
            data = json.loads(url.read().decode())
            for obj in data:
                symbol = obj["symbol"]
                if(obj["exchangeSuffix"] and obj["exchangeSuffix"] != "" and obj["exchangeSuffix"] in symbol):
                    symbol = symbol[:symbol.index(obj["exchangeSuffix"])]
                csv_writer.writerow([symbol, obj["name"], obj["region"], obj["currency"], obj['exchangeName']])
            print('finished writing symbols for {}'.format(region))
    
print('Finished writing symbols')
```
