```python
import pandas as pd
import os
```

```python
INPUT_DATA_LOC = "input/all_hist_data"
```

```python
def get_symbols():    
    return [x.split(".")[0] for x in os.listdir(INPUT_DATA_LOC)]
get_stock_list = get_symbols
get_stonks = get_symbols
len(get_symbols())
```

```python
def get_symbol_history(symbol):
    if os.path.exists(os.path.join(INPUT_DATA_LOC, symbol+".csv")):
        return pd.read_csv(os.path.join(INPUT_DATA_LOC, symbol+".csv"))
```

```python
df_refex = get_symbol_history("TI")
df_refex.sort_values("Timestamp")
df_refex.head(n=20)
```

```python
df_refex["ma_10"] = df_refex["Close Price"].rolling(window=10).mean()
df_refex
```
